```python
# Step 1: Load and inspect data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("NVDA_alpha.csv")

# Display first few rows
print("Preview of dataset:")
print(df.head())

# Info about data types and missing values
print("\nData Info:")
print(df.info())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Basic stats summary
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Step 2: Handle missing data
df = df.dropna()  # drop rows with missing values for simplicity

# Add new features
df["Price_Change"] = ((df["Close"] - df["Open"]) / df["Open"]) * 100
df["Direction"] = np.where(df["Price_Change"] > 0, 1, 0)

# Convert 'date' to datetime and sort
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

# Reset index
df.reset_index(drop=True, inplace=True)

# Display updated data
print("\nUpdated dataset with new features:")
print(df.head())
```

```python
# Step 3: Check correlation and visualize
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()

print("\nFinal columns in dataset:")
print(df.columns.tolist())
```

```python
corr = df.corr()
corr_price = corr['Price_Change'].sort_values(ascending=False)
corr_direction = corr['Direction'].sort_values(ascending=False)
```

```python
print("\n📈 Top correlations with Price_Change:\n")
print(corr_price)

print("\n🎯 Top correlations with Direction:\n")
print(corr_direction)

# Optional: Visualize as bar plots
plt.figure(figsize=(10,5))
sns.barplot(x=corr_price.index, y=corr_price.values, palette="coolwarm")
plt.title("Feature Correlation with Price Change")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(x=corr_direction.index, y=corr_direction.values, palette="coolwarm")
plt.title("Feature Correlation with Direction")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ==============================================
# 📊 Add Missing Features to Dataset
# ==============================================
import pandas as pd
import numpy as np

print("="*60)
print("📥 Loading Dataset...")
print("="*60)

# Load dataset
df = pd.read_csv("NVDA_dataset.csv")
print(f"\n📋 Original Dataset Shape: {df.shape}")
print(f"📋 Current Columns: {df.columns.tolist()}\n")

# ==========================================
# 1️⃣ Check and Add Missing Features
# ==========================================
print("🔧 Adding Missing Features...\n")

# 1. SMA5 (Simple Moving Average - 5 days)
if 'SMA5' not in df.columns:
    print("   ✅ Adding SMA5...")
    df['SMA5'] = df['Close'].rolling(window=5).mean()
else:
    print("   ℹ️ SMA5 already exists")

# 2. SMA20 (Simple Moving Average - 20 days)
if 'SMA20' not in df.columns:
    print("   ✅ Adding SMA20...")
    df['SMA20'] = df['Close'].rolling(window=20).mean()
else:
    print("   ℹ️ SMA20 already exists (as SMA_20)")

# 3. SMA50 (Simple Moving Average - 50 days)
if 'SMA50' not in df.columns:
    print("   ✅ Adding SMA50...")
    df['SMA50'] = df['Close'].rolling(window=50).mean()
else:
    print("   ℹ️ SMA50 already exists")

# 4. Price_Change (Percentage change)
if 'Price_Change' not in df.columns:
    print("   ✅ Adding Price_Change...")
    df['Price_Change'] = df['Close'].pct_change()
else:
    print("   ℹ️ Price_Change already exists")

# 5. RSI (Relative Strength Index)
if 'RSI' not in df.columns:
    print("   ✅ Calculating RSI...")
    def calculate_rsi(data, period=14):
        delta = data.diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
        rs = gain / loss
        rsi = 100 - (100 / (1 + rs))
        return rsi
    
    df['RSI'] = calculate_rsi(df['Close'], period=14)
else:
    print("   ℹ️ RSI already exists")

# 6. MACD (Moving Average Convergence Divergence)
if 'MACD' not in df.columns:
    print("   ✅ Calculating MACD...")
    ema_12 = df['Close'].ewm(span=12, adjust=False).mean()
    ema_26 = df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD'] = ema_12 - ema_26
else:
    print("   ℹ️ MACD already exists")

# 7. MACD_Signal (Signal line for MACD)
if 'MACD_Signal' not in df.columns:
    print("   ✅ Calculating MACD_Signal...")
    if 'MACD' in df.columns:
        df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
    else:
        ema_12 = df['Close'].ewm(span=12, adjust=False).mean()
        ema_26 = df['Close'].ewm(span=26, adjust=False).mean()
        macd = ema_12 - ema_26
        df['MACD_Signal'] = macd.ewm(span=9, adjust=False).mean()
else:
    print("   ℹ️ MACD_Signal already exists")

print("\n" + "="*60)
print("✅ All Features Added Successfully!")
print("="*60)

# ==========================================
# 2️⃣ Remove NaN values (created by rolling calculations)
# ==========================================
print(f"\n📊 Handling NaN values...")
print(f"   Before: {df.shape[0]} rows")
df_clean = df.dropna()
print(f"   After: {df_clean.shape[0]} rows")
print(f"   Removed: {df.shape[0] - df_clean.shape[0]} rows\n")

# ==========================================
# 3️⃣ Display New Dataset Info
# ==========================================
print("="*60)
print("📋 Final Dataset Information")
print("="*60)
print(f"\n✅ Final Shape: {df_clean.shape}")
print(f"✅ Total Columns: {len(df_clean.columns)}")
print(f"\n📌 Columns in Dataset:")
for i, col in enumerate(df_clean.columns, 1):
    print(f"   {i}. {col}")

print(f"\n📊 Dataset Preview (first 5 rows):")
print(df_clean.head())

print(f"\n📊 Dataset Statistics:")
print(df_clean.describe())

# ==========================================
# 4️⃣ Verify Required Features
# ==========================================
print("\n" + "="*60)
print("🎯 Checking Required Features")
print("="*60)

required_features = ['Close', 'SMA5', 'SMA20', 'SMA50', 
                     'Price_Change', 'RSI', 'MACD', 'MACD_Signal']

print("\nRequired features for model training:")
for feature in required_features:
    if feature in df_clean.columns:
        print(f"   ✅ {feature}")
    else:
        print(f"   ❌ {feature} - MISSING!")

# ==========================================
# 5️⃣ Save Updated Dataset
# ==========================================
print("\n" + "="*60)
print("💾 Saving Updated Dataset")
print("="*60)

output_file = "NVDA_dataset_updated.csv"
df_clean.to_csv(output_file, index=False)
print(f"\n✅ Dataset saved as: {output_file}")
print(f"✅ Ready for model training!\n")

# ==========================================
# 6️⃣ Create Backup
# ==========================================
backup_file = "NVDA_dataset_backup.csv"
df.to_csv(backup_file, index=False)
print(f"💾 Backup created: {backup_file}")
print(f"   (Original dataset with NaN values)\n")

print("="*60)
print("✨ Feature Engineering Complete!")
print("="*60)
print("\n📝 Next Steps:")
print("   1. Use 'NVDA_dataset_updated.csv' for training")
print("   2. Update your training script to use the new dataset")
print("   3. Run the training script to build the improved model\n")
```
