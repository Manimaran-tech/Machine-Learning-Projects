```python
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------
# 1️⃣ Load data
# -------------------------
stocks = ["NVDA"]
data = {}

for stock in stocks:
    df = pd.read_csv(f"{stock}_alpha.csv", parse_dates=['date'], index_col='date')
    data[stock] = df

# -------------------------
# 2️⃣ Plot Close Price with SMA & EMA
# -------------------------
plt.figure(figsize=(16,12))
for i, stock in enumerate(stocks, 1):
    plt.subplot(4,1,i)
    df = data[stock]
    plt.plot(df['Close'], label='Close', color='blue')
    plt.plot(df['SMA_20'], label='SMA 20', color='orange')
    plt.plot(df['EMA_50'], label='EMA 50', color='green')
    plt.title(f"{stock} Price with SMA_20 & EMA_50")
    plt.ylabel("Price ($)")
    plt.legend()
plt.xlabel("Date")
plt.tight_layout()
plt.show()

# -------------------------
# 3️⃣ Plot RSI
# -------------------------
plt.figure(figsize=(16,12))
for i, stock in enumerate(stocks, 1):
    plt.subplot(4,1,i)
    df = data[stock]
    plt.plot(df['RSI'], label='RSI', color='purple')
    plt.axhline(70, color='red', linestyle='--', alpha=0.5)
    plt.axhline(30, color='green', linestyle='--', alpha=0.5)
    plt.title(f"{stock} RSI")
    plt.ylabel("RSI")
    plt.legend()
plt.xlabel("Date")
plt.tight_layout()
plt.show()

# -------------------------
# 4️⃣ Plot MACD
# -------------------------
plt.figure(figsize=(16,12))
for i, stock in enumerate(stocks, 1):
    plt.subplot(4,1,i)
    df = data[stock]
    plt.plot(df['MACD'], label='MACD', color='brown')
    plt.title(f"{stock} MACD")
    plt.ylabel("MACD")
    plt.legend()
plt.xlabel("Date")
plt.tight_layout()
plt.show()

# -------------------------
# 5️⃣ Plot Bollinger Bands
# -------------------------
plt.figure(figsize=(16,12))
for i, stock in enumerate(stocks, 1):
    plt.subplot(4,1,i)
    df = data[stock]
    plt.plot(df['Close'], label='Close', color='blue')
    plt.plot(df['BB_upper'], label='BB Upper', color='red', linestyle='--')
    plt.plot(df['BB_middle'], label='BB Middle', color='green', linestyle='--')
    plt.plot(df['BB_lower'], label='BB Lower', color='red', linestyle='--')
    plt.fill_between(df.index, df['BB_lower'], df['BB_upper'], color='gray', alpha=0.1)
    plt.title(f"{stock} Bollinger Bands")
    plt.ylabel("Price ($)")
    plt.legend()
plt.xlabel("Date")
plt.tight_layout()
plt.show()

# -------------------------
# 6️⃣ Plot Volume
# -------------------------
plt.figure(figsize=(16,12))
for i, stock in enumerate(stocks, 1):
    plt.subplot(4,1,i)
    df = data[stock]
    plt.bar(df.index, df['Volume'], color='skyblue')
    plt.title(f"{stock} Daily Volume")
    plt.ylabel("Volume")
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```
